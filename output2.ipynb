```python
import pandas as pd
grocery = pd.read_csv("/content/main1.csv")
grocery.head()
```

```python
grocery["sales_amount"]=grocery["sales_quantity"].where(grocery["unit"]!="pcs",
    grocery["product_description"].str.split("-",expand=True)[1].astype("float")*\
    grocery["sales_quantity"])
```

```python
grocery["price_new"] = grocery['price'].fillna(
   grocery.groupby('product_description')['price'].transform("mean")
)
```
